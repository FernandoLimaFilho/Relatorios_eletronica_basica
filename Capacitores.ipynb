#Carga e descarga de um capacitor

```python
"""
1) Importao do numpy para trabalhar com arrays
"""
import numpy as np
"""
2) Importao do matplotlib.pyplot para trabalhar com grficos
"""
import matplotlib.pyplot as plt
"""
3) De matplotlib.font_manager vamos importar FontProperties para criar fontes de texto.
"""
from matplotlib.font_manager import FontProperties

# Parmetros do circuito
R = 10e3         # resistncia em ohms (10 k)
C = 100e-6       # capacitncia em farads (100 F)
V0 = 5           # tenso inicial em volts

# Constante de tempo
tau = R * C

# Vetor de tempo: de 0 a 5 com 1000 pontos
t = np.linspace(0, 5*tau, 1000)

# Equao da descarga: V(t) = V0 * exp(-t/RC)
V_t = V0 * np.exp(-t / tau)

# Equao da carga: V(t) = V0 * (1- exp(-t/RC))
V_t_c = V0- V0 * np.exp(-t / tau)

# Criando pontos para as constantes de tempo (1, 2, ..., 5)
taus = [i * tau for i in range(1, 6)]
v_taus = [V0 * np.exp(-i) for i in range(1, 6)]

# Plotando o grfico
plt.figure(figsize=(10, 6))
plt.plot(t, V_t, label=r'$V(t) = V_0 \cdot e^{-\dfrac{t}{\tau}}$', color='blue')  # Equao formatada com LaTeX
plt.plot(t, V_t_c, label=r'$V(t) = V_0 \cdot \left(1- e^{-\dfrac{t}{\tau}}\right)$', color='green')  # Equao formatada com LaTeX

# Marcando as constantes de tempo no grfico
for i, (t_mark, v_mark) in enumerate(zip(taus, v_taus), start=1):
    plt.plot(t_mark, v_mark, 'ro')  # ponto vermelho
    plt.text(t_mark, v_mark + 0.2, f'{i}', ha='center', color='red')

#Criao das fontes
Fonte_1 = {'family': 'serif', 'color': 'black', 'weight': 'normal', 'size': 12}
Fonte_2 = FontProperties(family="serif", weight='normal', style='normal', size=16)

#Plotando o grfico
plt.axhline(0, color='gray', linestyle='--')
plt.title('Carga e descarga de um capacitor', fontdict=Fonte_1)
plt.xlabel('Tempo (s)', fontdict=Fonte_1)
plt.ylabel('Tenso no capacitor (V)', fontdict=Fonte_1)
plt.grid(True)
plt.legend(fontsize = "large")
plt.tight_layout()
plt.show()
```

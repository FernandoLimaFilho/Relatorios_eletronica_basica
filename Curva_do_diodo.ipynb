###üìå O que √© um Diodo?

Um diodo √© um componente eletr√¥nico semicondutor que permite a passagem de corrente el√©trica em apenas um sentido ‚Äî do √¢nodo (+) para o c√°todo (‚àí). Ele se comporta como uma v√°lvula de corrente.

###‚öôÔ∏è Funcionamento B√°sico:

Quando o diodo est√° polarizado diretamente (√¢nodo com potencial maior que o c√°todo), ele conduz corrente.

Quando est√° polarizado reversamente, ele bloqueia a corrente (exceto uma corrente de fuga muito pequena).

A tens√£o m√≠nima para que ele conduza (polariza√ß√£o direta) √© chamada de tens√£o de limiar ou tens√£o de joelho. Por exemplo:

Diodo de sil√≠cio: ~0,7 V

Diodo de germ√¢nio: ~0,3 V

###Dados coletados no laborat√≥rio (polariza√ß√£o direta)

```python
Vcc = [0, 0.1, 0.3, 0.5, 0.7, 1.0, 3.0, 5.0, 10.0, 15.0] #Tens√£o na fonte
Vd = [0.0, 0.07, 0.28, 0.46, 0.48, 0.53, 0.60, 0.63, 0.66, 0.68] #Tens√£o medida entre os polos do diodo
Id = [0.00, 0.00, 0.00, 0.06, 0.16, 0.31, 2.20, 3.80, 8.40, 12.95] # Corrente medida com o circuito aberto, entre o diodo e o resistor.
```

###Cria√ß√£o do dataframe

Importa√ß√£o das bibliotecas

```python
import pandas as pd
import numpy as np
```

Dataframe

```python
Dados_diodo_polarizado_diretamente = pd.DataFrame({'Vcc': Vcc, 'Vd': Vd, 'Id': Id})
Dados_diodo_polarizado_diretamente
```

###Curva do diodo diretamente polarizado (Id x Vd)

```python
#Importa√ß√£o das bibliotecas para gr√°ficos
import matplotlib.pyplot as plt
"""
1¬∞) De matplotlib.ticker vamos importar o AutoMinorLocator e o MaxNLocator para trabalhar com os "ticks"
    dos gr√°ficos.
"""
import matplotlib.ticker as mticker
from matplotlib.ticker import AutoMinorLocator, MaxNLocator
"""
2¬∞) De matplotlib.font_manager vamos importar FontProperties para criar fontes de texto.
"""
from matplotlib.font_manager import FontProperties
```

```python
#Cria√ß√£o das fontes
Fonte_1 = {'family': 'serif', 'color': 'black', 'weight': 'normal', 'size': 16}
Fonte_2 = FontProperties(family="serif", weight='normal', style='normal', size=16)
#Aloca√ß√£o da figura
fig, ax = plt.subplots(figsize = (7,7))
#Plot
plt.plot(Dados_diodo_polarizado_diretamente["Vd"], Dados_diodo_polarizado_diretamente["Id"], color = "gray", linewidth=3)
plt.xlabel("Vd (V)", fontdict = Fonte_1)
plt.ylabel("Id (mA)", fontdict = Fonte_1)
#T√≠tulo
plt.title("Curva do diodo", fontdict = Fonte_1)
#Ticks nos eixos
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.tick_params(axis = "both", direction = "in", labelcolor='black', labelsize=16, top = True, right = True)
ax.tick_params(which='major', direction = "in", color='black', length=7, width = 3)
ax.tick_params(which='minor', direction = "in", length=4, color='black', width = 2, top = True, right = True)
#Cor dos eixos
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_color("black")
for axis in ["top", "bottom", "left", "right"]:
    ax.spines[axis].set_linewidth(1.3)
#Mostrar
plt.show()
```

#Carga e descarga de um capacitor

## Carga do Capacitor

Ao conectar um capacitor a uma fonte de tensão através de um resistor, ele se carrega gradualmente. A carga Q(t) no capacitor ao longo do tempo é dada por:

$$
Q(t) = Q_{\text{máx}} \left(1 - e^{-\frac{t}{RC}}\right)
$$

A tensão \( V(t) \) nos terminais do capacitor cresce de forma similar:

$$
V(t) = V_{\text{f}} \left(1 - e^{-\frac{t}{RC}}\right)
$$

**Onde:**
- $Q_{\text{máx}}$ é a carga máxima,
- $V_{\text{f}}$ é a tensão da fonte,
- $R$ é a resistência,
- $C$ é a capacitância,
- $\tau = RC$ é a constante de tempo.

Durante a carga:
- No início t=0, a carga e a tensão são nulas.
- Com o tempo, ambos se aproximam dos valores máximos de forma exponencial.

## Descarga do Capacitor

Ao desconectar a fonte e deixar o capacitor ligado apenas ao resistor, ele se descarrega. A carga \( Q(t) \) ao longo do tempo é:

$$
Q(t) = Q_0 \, e^{-\frac{t}{RC}}
$$

E a tensão \( V(t) \) decresce da mesma maneira:

$$
V(t) = V_0 \, e^{-\frac{t}{RC}}
$$

**Onde:**
- $Q_0$ e $V_0$ são os valores iniciais de carga e tensão.

Durante a descarga:
- A carga e a tensão diminuem exponencialmente.
- Após um tempo de aproximadamente $5\tau$, o capacitor é praticamente considerado descarregado.



```python
"""
1°) Importação do numpy para trabalhar com arrays
"""
import numpy as np
"""
2°) Importação do matplotlib.pyplot para trabalhar com gráficos
"""
import matplotlib.pyplot as plt
"""
3°) De matplotlib.font_manager vamos importar FontProperties para criar fontes de texto.
"""
from matplotlib.font_manager import FontProperties

# Parâmetros do circuito
R = 10e3         # resistência em ohms (10 kΩ)
C = 100e-6       # capacitância em farads (100 µF)
V0 = 5           # tensão inicial em volts

# Constante de tempo
tau = R * C

# Vetor de tempo: de 0 a 5τ com 1000 pontos
t = np.linspace(0, 5*tau, 1000)

# Equação da descarga: V(t) = V0 * exp(-t/RC)
V_t = V0 * np.exp(-t / tau)

# Equação da carga: V(t) = V0 * (1- exp(-t/RC))
V_t_c = V0- V0 * np.exp(-t / tau)

# Criando pontos para as constantes de tempo (1τ, 2τ, ..., 5τ)
taus = [i * tau for i in range(1, 6)]
v_taus = [V0 * np.exp(-i) for i in range(1, 6)]

# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(t, V_t, label=r'$V(t) = V_0 \cdot e^{-\dfrac{t}{\tau}}$', color='blue')  # Equação formatada com LaTeX
plt.plot(t, V_t_c, label=r'$V(t) = V_0 \cdot \left(1- e^{-\dfrac{t}{\tau}}\right)$', color='green')  # Equação formatada com LaTeX

# Marcando as constantes de tempo no gráfico
for i, (t_mark, v_mark) in enumerate(zip(taus, v_taus), start=1):
    plt.plot(t_mark, v_mark, 'ro')  # ponto vermelho
    plt.text(t_mark, v_mark + 0.2, f'{i}τ', ha='center', color='red')

#Criação das fontes
Fonte_1 = {'family': 'serif', 'color': 'black', 'weight': 'normal', 'size': 12}
Fonte_2 = FontProperties(family="serif", weight='normal', style='normal', size=16)

#Plotando o gráfico
plt.axhline(0, color='gray', linestyle='--')
plt.title('Carga e descarga de um capacitor', fontdict=Fonte_1)
plt.xlabel('Tempo (s)', fontdict=Fonte_1)
plt.ylabel('Tensão no capacitor (V)', fontdict=Fonte_1)
plt.grid(True)
plt.legend(fontsize = "large")
plt.tight_layout()
plt.show()
```

# Carga e Descarga de um Indutor

## Estabelecimento da Corrente (Carga)

Ao conectar um indutor a uma fonte de tensão através de um resistor, a corrente $i(t)$ aumenta de forma exponencial:

$$
i(t) = i_{\text{máx}} \left(1 - e^{-\frac{R}{L}t}\right)
$$

A tensão no indutor, por outro lado, decresce:

$$
V_L(t) = V \, e^{-\frac{R}{L}t}
$$

**Onde:**
- $i_{\text{máx}} = \frac{V}{R}$ é a corrente máxima,
- $V$ é a tensão da fonte,
- $R$ é a resistência,
- $L$ é a indutância,
- $\tau = \frac{L}{R}$ é a constante de tempo do circuito.

Durante o estabelecimento da corrente:
- Inicialmente t=0, a corrente é zero e a tensão no indutor é máxima.
- Com o tempo, a corrente cresce até atingir $i_{\text{máx}}$ e a tensão no indutor cai para zero.

## Desligamento da Corrente (Descarga)

Ao desligar a fonte, o indutor tenta manter a corrente fluindo. A corrente decresce de forma exponencial:

$$
i(t) = i_0 \, e^{-\frac{R}{L}t}
$$

**Onde:**
- $i_0$ é a corrente inicial no momento do desligamento.

Durante a descarga:
- A corrente diminui exponencialmente.
- Após um tempo de aproximadamente $5\tau$, a corrente é considerada nula.



```python
"""
Constante de tempo do indutor
L = Indutância
R = Resistência
"""
L = 1200
R = 500
V0 = 5
tau_indutor = L/R

#Equação de carga
Vi_t_carga = V0*np.exp(-t/tau_indutor)

#Equação de descarga
Vi_t_descarga = V0*(1-np.exp(-t/tau_indutor))

# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(t, Vi_t_carga, label=r'$V(t) = V_0 \cdot e^{-\dfrac{t}{\tau}}$', color='blue')  # Equação formatada com LaTeX
plt.plot(t, Vi_t_descarga, label=r'$V(t) = V_0 \cdot \left(1- e^{-\dfrac{t}{\tau}}\right)$', color='green')  # Equação formatada com LaTeX

#Criação das fontes
Fonte_1 = {'family': 'serif', 'color': 'black', 'weight': 'normal', 'size': 12}
Fonte_2 = FontProperties(family="serif", weight='normal', style='normal', size=16)

#Plotando o gráfico
plt.axhline(0, color='gray', linestyle='--')
plt.title('Carga e descarga de um indutor', fontdict=Fonte_1)
plt.xlabel('Tempo (s)', fontdict=Fonte_1)
plt.ylabel('Tensão no indutor (V)', fontdict=Fonte_1)
plt.grid(True)
plt.legend(fontsize = "large")
plt.tight_layout()
plt.show()
```
